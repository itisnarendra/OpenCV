```python
import requests

url_base = 'https://facultyweb.centennialcollege.ca/npersh/workshop/python/'
img_file = 'stage.png'

# response = requests.get(f'{url_base}{img_file}')
try:
    response = requests.get(f'{url_base}{img_file}')
    response.raise_for_status()  # Raise an exception for bad status codes

    # If it's a binary file like an image or a zip file:
    file_content = response.content
    with open(img_file, 'wb') as f:
        f.write(file_content)
    print(f'File downloaded successfully as {img_file}')

except requests.exceptions.RequestException as e:
    print(f'Error fetching the file: {e}')
except Exception as e:
    print(f'An unexpected error occurred: {e}')
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('stage.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
